```python
X = market[['overall_qual', 'overall_cond', 'Total Bath', 'fireplaces', 'ms_subclass', 'ms_zoning', 'lot_area', 'street', 'lot_shape', 'neighborhood', 'house_style', 'year_built', 'bedroom_abvgr', 'garage_type', 'garage_cars', 'gr_liv_area', 'bsmt_cond', 'land_slope', 'roof_matl', 'exter_qual', 'exter_cond', 'foundation', 'heating', 'central_air', 'sale_type', 'mo_sold', 'yr_sold', 'totrms_abvgrd', 'wood_deck_sf','total_sq', 'full_bath', 'half_bath', 'bsmt_qual', 'kitchen_qual']]
y = market[['saleprice']]

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
ct = ColumnTransformer([
    ('pl', Pipeline([
        ('si', SimpleImputer()),
        ('sc', StandardScaler()),
        ('pf', PolynomialFeatures(degree=2, include_bias=False))]),['overall_qual', 'overall_cond', 'Total Bath', 'fireplaces', 'gr_liv_area', 'bedroom_abvgr', 'garage_cars', 'lot_area', 'totrms_abvgrd', 'wood_deck_sf', 'total_sq', 'full_bath', 'half_bath']),
        ('oh', OneHotEncoder(drop='first', sparse_output=False, handle_unknown='ignore'), ["ms_subclass", 'neighborhood', 'house_style', 'garage_type', 'bsmt_cond', 'ms_zoning', 'street', 'lot_shape', 'land_slope', 'roof_matl', 'exter_qual', 'exter_cond', 'foundation', 'heating', 'central_air', 'sale_type', 'kitchen_qual', 'bsmt_qual'])
])
```

```python
X_train_ct = ct.fit_transform(X_train)
X_test_ct = ct.transform(X_test)
```

```python
lr = LinearRegression()
lr.fit(X_train_ct, y_train)
```

```python
lr.score(X_train_ct, y_train)
```

```python
lr.score(X_test_ct, y_test)
```

```python
y_pred = lr.predict(X_test_ct)
```

```python
metrics.r2_score(y_test, y_pred)
```

```python
ridge_model = Ridge(alpha=10)
ridge_model.fit(X_train_ct, y_train)
r_alphas = np.logspace(0, 5, 100)

ridge_cv = RidgeCV(alphas=r_alphas, scoring='r2', cv=5)

ridge_cv = ridge_model.fit(X_train_ct, y_train)
```

```python
ridge_cv.alpha
```

```python
print(ridge_cv.score(X_train_ct, y_train))
print(ridge_cv.score(X_test_ct, y_test))
```

```python
l_alphas = np.logspace(-3, 0, 100)

lasso_cv = LassoCV(alphas=l_alphas, cv=5, max_iter=50000)

lasso_cv.fit(X_train_ct, y_train);
```
